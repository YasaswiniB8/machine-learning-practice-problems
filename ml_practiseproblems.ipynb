```python
# 3rd problem
import math
def bidi(n,p,x):
    b=((math.factorial(n))/(math.factorial(x)*math.factorial(n-x)))*(p**x)*((1-p)**(n-x))
    return b
print(bidi(10,0.5,6))
print(bidi(10,2/3,6))
print(bidi(10,1/3,6))
```
